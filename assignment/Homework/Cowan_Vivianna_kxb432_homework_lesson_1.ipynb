### 1.3 Basic Arithmetic

R can work like a calculator! Here are the basic math operations:

## Part 2: Data Structures - Vectors and Data Frames

Now let's learn about how R stores groups of data. This is essential for working with real datasets!

### 2.1 Creating Vectors

A **vector** is a list of values of the same type. We create vectors using the `c()` function (c stands for "combine").

**üìù Task:** Run the cell below to create and view vectors:

````r
# Example: Creating vectors

# A vector of numbers
prices = c(10.99, 25.50, 8.75, 15.00, 32.99)
print("Prices:")
print(prices)

# A vector of text (character strings)
products = c("Apple", "Banana", "Orange", "Grape", "Mango")
print("Products:")
print(products)
````

### 2.3 Understanding Data Frames

A **data frame** is like a spreadsheet - it has rows and columns. Each column can contain different types of data. This is the most important data structure for data analysis!

**üìù Task:** Run this cell to create a simple data frame:

````r
# Example: Creating a data frame
employee_data = data.frame(
  Name = c("Alice", "Bob", "Carol"),
  Age = c(28, 35, 42),
  Salary = c(55000, 62000, 78000)
)

# Print the data frame
print("Employee Data:")
print(employee_data)
````

### 3.2 Understanding the Working Directory

The **working directory** is the folder where R looks for files and saves files. Think of it as R's "current location" on your computer.

**Key functions:**
| Function | What It Does | Example |
|----------|--------------|---------|
| `getwd()` | **Get** working directory - shows current location | `getwd()` |
| `setwd()` | **Set** working directory - changes location | `setwd("/path/to/folder")` |

**Why this matters:**
- If R can't find your file, it might be looking in the wrong folder
- Use `getwd()` to see where R is looking
- Use `setwd()` to change to the correct folder

**üìù Task:** Write the code to check your current working directory:

## Part 4: Inspecting Your Data

Once you've loaded data, you need to explore it! Here are the essential functions:
- `head()` - shows the first few rows
- `str()` - shows the structure and data types
- `summary()` - shows statistical summaries

### 4.1 View the First Few Rows

The `head()` function shows you the first few rows of your data (default is 6 rows).

**Syntax:**
```r
head(data_frame_name)        # Shows first 6 rows
head(data_frame_name, 10)    # Shows first 10 rows
```

**üìù Task:** Write the code to display the first 10 rows of `sales_data`:

````r
# YOUR CODE HERE
# Use head() to view the first 10 rows of sales_data
# Hint: head(data_name, number_of_rows)

head(sales_data, 10)
````

### 4.2 Examine the Data Structure

The `str()` function (short for "structure") tells you:
- How many rows and columns
- The name of each column
- The data type of each column (num = number, chr = text)

**Syntax:**
```r
str(data_frame_name)
```

**üìù Task:** Write the code to view the structure of `sales_data`:

````r
# YOUR CODE HERE
# Use str() to view the structure of sales_data

str(sales_data)
````

### 4.3 Get Summary Statistics

The `summary()` function gives you statistics for each column:
- For numbers: min, max, mean, median, quartiles
- For text: length, class, mode

**Syntax:**
```r
summary(data_frame_name)
```

**üìù Task:** Write the code to get summary statistics for `sales_data`:

````r
# YOUR CODE HERE
# Use summary() to get summary statistics for sales_data

summary(sales_data)
````
